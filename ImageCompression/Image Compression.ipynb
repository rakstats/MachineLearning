```python
from matplotlib.image import imread
import matplotlib.pyplot as plt
import numpy as np
import os

plt.rcParams['figure.figsize'] = [10, 10]

image_rgb = imread('Ruchira.jpg')
image_gray = np.mean(image_rgb, -1)

img = plt.imshow(image_gray)
img.set_cmap('gray')
plt.axis('off')
plt.show()
```

```python
U, S, VT = np.linalg.svd(image_gray, full_matrices = False)
S = np.diag(S)

j=0
for r in (10, 50, 100):
    image_approx = U[:, :r] @ S[0:r, :r] @ VT[:r, :]
    plt.figure(j+1)
    j += 1
    img = plt.imshow(image_approx)
    img.set_cmap('gray')
    plt.axis('off')
    plt.title('r = ' + str(r))
    plt.show()
```

```python
plt.figure(1)
plt.semilogy(np.diag(S))
plt.title('Singular Values')
plt.show()

plt.figure(2)
plt.plot(np.cumsum(np.diag(S))/np.sum(np.diag(S)))
plt.title('Singular Values: Cumulative Sum')
plt.show()
```
